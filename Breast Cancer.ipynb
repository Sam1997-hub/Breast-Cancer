```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('breast_data.csv')
```

```python
df.head()
```

```python
df.drop(['id'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(round(df.corr(),2),annot=True)
```

```python
X=df.drop(['diagnosis'],axis=1)
y=df['diagnosis']
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
var=VarianceThreshold(threshold=0.25)
var.fit_transform(X)
```

```python
len(var.feature_names_in_)
```

```python
var.get_support()
```

```python
dropcols=[column for column in X.columns
          if column not in X.columns[var.get_support()]]
dropcols
```

```python
X.drop(dropcols,axis=1,inplace=True)
```

```python
X.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
y=le.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
X_train_scaled=ss.fit_transform(X_train)
```

```python
X_test_scaled=ss.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegressionCV
```

```python
lr=LogisticRegressionCV(cv=20)
```

```python
lr.fit(X_train_scaled,y_train)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
y_train_pred=lr.predict(X_train_scaled)
```

```python
sns.heatmap(confusion_matrix(y_train,y_train_pred),annot=True,fmt='0.5g',cmap='Blues')
print(classification_report(y_train,y_train_pred))
```

```python
y_test_pred=lr.predict(X_test_scaled)
```

```python
sns.heatmap(confusion_matrix(y_test,y_test_pred),annot=True,fmt='0.5g',cmap='Blues')
print(classification_report(y_test,y_test_pred))
```

```python
from sklearn.metrics import roc_auc_score,roc_curve
```

```python
pred_prob=lr.predict_proba(X_test_scaled)
pred_prob=pred_prob[:,1]
print(roc_auc_score(y_test,pred_prob))
fpr, tpr, thresholds =roc_curve(y_test,pred_prob)
plt.plot(fpr,tpr,marker='o',linestyle='--')
plt.grid()
plt.show()
```
